```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
url = 'https://www.scrapethissite.com/pages/ajax-javascript/#2015'
driver = webdriver.Chrome()
driver.get(url)

driver.implicitly_wait(10)

#Find all the rows
films = driver.find_elements(By.XPATH, '//*[@id="table-body"]/tr')

film_list_global = []

# Iterate through each film row and extract details
for film in films:
    film_dict = {}
    film_dict['title'] = film.find_element(By.XPATH, './td[1]').text
    film_dict['nominations'] = film.find_element(By.XPATH, './td[2]').text
    film_dict['awards'] = film.find_element(By.XPATH, './td[3]').text
    film_list_global.append(film_dict)

# Close the browser
driver.quit()

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(film_list_global)

# Save the DataFrame to CSV
df.to_csv('films_2015.csv')

# Print the DataFrame
print(df)
```
